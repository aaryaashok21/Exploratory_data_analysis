# Description

### Analyzing,vizualising and cleaning a dataset for customer segementation ,understanding the relations and trends between various products. Dataset contains 55000 entries with 13 columns regarding customer demographics,product details,purchasing and payment methods etc. 

## Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("all imported")
```

# Loading data set  

### Functions used:
### load data-read_csv, head()- to load first 10 rows, tail()-to load last 10 rows

```python
df=pd.read_csv('E:\data science\dataset\ecommerce data for data analysis.csv')
print(df.head(10))
print(df.tail(10))
```

## Get info of dataset 
### Function used : info()

```python
print(df.info())
```

## Checking null values 
### functions used : isna(),sum()

```python
print(df.isna().sum())
```

## Deleting null value cells
### function used: dropna(),isna(),sum()

```python
df.dropna(inplace=True)
print(df.isna().sum())
```

## Checking duplicate values
### function used : duplicated(),sum()

```python
print(df.duplicated().sum())
```

## renaming the column names
### function used: rename()

```python
print(df.rename(columns={'CID':'customerid','TID':'transcid','Gender':'gender','Age Group':'agegroup','Purchase Date':'purchasedate',
                  'Product Category':'prodcategory','Discount Availed':'discavailed','Discount Name':'discname','Discount Amount (INR)':'discamt',
                  'Gross Amount':'grossamt','Net Amount':'netamt','Purchase Method':'purchmethod','Location':'location'},inplace=True))
print(df)
```

## Changing datatype
### function used : astype()

```python
df['gender']=df['gender'].astype('string')
df['prodcategory']=df['prodcategory'].astype('string')
df['discavailed']=df['discavailed'].astype('string')
df['discname']=df['discname'].astype('string')
df['purchmethod']=df['purchmethod'].astype('string')
df['location']=df['location'].astype('string')
```

## converting date format 
### function used : to_datetime() 

```python
df['purchasedate'] = pd.to_datetime(df['purchasedate'])
print(df['purchasedate'])
```

## Finding outliers using boxplot in seaborn

```python
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

### No outliers are present 

## Statistical analysis 
### functions used: describe(),cov(),corr(),std()

```python
print(df.describe().round())
print('correlation')
print(df.corr(numeric_only=True).__round__())
print("covariance")
print(df.cov(numeric_only=True).__round__())
print("standard deviation")
print(df.std(numeric_only=True))
```

# EXPLORATORY DATA ANALYSIS
## Questions for analyzing the dataset

1.which product has been purchased more by customers

```python
print(df.groupby('prodcategory').size().sort_values(ascending=False))
P1=df.groupby('prodcategory').size()
print(P1.sort_values(ascending=False).head(1))
P1.plot(kind='bar',title='highest sold product')
plt.xlabel('products')
plt.ylabel('number of purchases')
plt.show()
```

2.Which gender has made more number of purchases

```python
g1=df.groupby('gender').size()
print(g1.sort_values(ascending=False).head(1))
g1.plot(kind='bar',title='Gender-Wise Distribution of Purchases')
plt.xlabel('gender')
plt.ylabel('number of purchases ')
plt.show()
```

3.Which mode of payment has been used more number of times

```python
pr1=df.groupby('purchmethod').size()
print(pr1.sort_values(ascending=False).head(1))
pr1.plot(kind='bar',title='Mode of payment wise distribution of purchases')
plt.xlabel('mode of payment')
plt.ylabel('number of purchases')
plt.show()
```

4.Count of each product sold

```python
pd1=df.groupby('prodcategory').size()
print(pd1)
pd1.plot(kind='bar',title='product category wise distribution')
plt.xlabel('products')
plt.ylabel('number products sold')
plt.show()
```

5.Which discount name has applied more number of times

```python
d1=df.groupby('discname').size()
print(d1.sort_values(ascending=False).head(1))
d1.plot(kind='bar',title='discount name wise distribution')
plt.xlabel('discount name')
plt.ylabel('umber of times applied')
plt.show()
```

6.Count of purchase methods adopted by customers

```python
pmd=df.groupby('purchmethod').size()
print(pmd)
pmd.plot(kind='bar',title='distribution according to purchase method')
plt.xlabel('purchase method')
plt.ylabel('count of purchase method')
plt.show()
```

7.average of each purchase methods 


```python
f1=df.groupby('purchmethod').size().sum()
print(f1)
s2=df.groupby('purchmethod').size()
print('AVERAGE:')
print('cash on delivery=',(s2['Cash on Delivery']/f1))
print('Credit Card',s2['Credit Card']/f1)
print('Google Pay UPI=',s2['Google Pay UPI']/f1)
print('International Card=',s2['International Card']/f1)
print('Net Banking=',s2['Net Banking']/f1)
print('Paytm UPI=',s2['Paytm UPI']/f1)
print('PhonePe UPI',s2['PhonePe UPI']/f1)
```

8.correlation between gross amount and net amount

9.correlation between discount amount and gross amount

```python
print(df['discamt'].corr(df['grossamt']).__round__())
```

10.correlation between discount amount and net amount

```python
print(df['discamt'].corr(df['netamt']).__round__())
```

11.Which age group has done more purchases

```python
agerp=df.groupby('agegroup').size()
print(agerp)
print('age group with highest purchases:')
print(agerp.sort_values(ascending=False).head(1))
agerp.plot(kind='bar',title='distribution based on age group')
plt.xlabel('age group')
plt.ylabel('purchases made')
plt.show()
```

12.Which location has done highest purchases

```python
a=df.groupby('location').size()
print(a)
print("location with large number of purchases:")
print(a.sort_values(ascending=False).head(1))
a.plot(kind='bar',title='number of purchases by location')
plt.xlabel('location')
plt.ylabel('number of purchases')
plt.show()
```

13.find the discount amount of each discount name

```python
dinm=df.groupby('discname')['discamt'].sum()
print(dinm)
dinm.plot(kind='bar',title='distribution based on discount name')
plt.xlabel('discount name')
plt.ylabel('count')
plt.show()
```

14.During which year highest purchases has been done by customers

```python

df['year']=df['purchasedate'].dt.year
t1=df.groupby('year').size()
print(t1.sort_values(ascending=False).head(1))
t1.plot(kind='bar',title='date wise distribution')
plt.xlabel("year")
plt.ylabel("no.of purchases")
plt.show()
```

15.what is the count of products sold per year

```python
print(df.groupby(['year','prodcategory']).size())
```

16.Find the product category with highest gross amount per year

```python
high_product=df.loc[df.groupby('year')['grossamt'].idxmax(),['year','prodcategory','grossamt']]
print(high_product)
high_product['year-product']=high_product['year'].astype('string')+ '-'+ high_product['prodcategory']
high_product.plot(x='year-product',y='grossamt',kind='bar',title='product category distribution based on gross amount')
plt.xlabel('year-product')
plt.ylabel('gross amount')
plt.xticks(rotation=50)
plt.show()
```

17.Total discount amount  per location 

```python
print(df.groupby('location')['discamt'].sum(),['location','discamt'])
ds=df.groupby('location')['discamt'].sum().sort_values(ascending=False).head(1)
print('location with highest discount amount is:')
print(ds)
```

18.average net amout spend based on age group

```python
print(df.groupby('agegroup')['netamt'].mean())
print(df.groupby('agegroup')['netamt'].mean().sort_values(ascending=False).head(1))
```

19.Analysis of discount availed

```python
discount_analyse=df['discavailed'].value_counts(normalize=True)*100
print(discount_analyse)
```

20.To find the year wise revenue trend

```python
trend=df.groupby("year")['netamt'].sum()
formateed=trend.apply(str)
print(formateed)
trend.plot(kind='line',marker='o',title='year wise revenue trend')
plt.xlabel('year')
plt.ylabel('net amount')
plt.show()
```
